```python
import pandas as pd
import random
from faker import Faker
from br_nome_gen import pessoa_random
```

```python
df = pd.DataFrame(columns=['Cliente', 'Produto', 'Categoria_Produto', 'Data_Compra', 'Quantidade', 'Preco'])

categorias_e_produtos = {
    'Eletrônicos': ['Computador', 'Celular', 'Smart TV', 'Headset', 'Videogames', 'Tablet', 'Monitor'],
    'Eletrodomésticos': ['Máquina de Café', 'Impressora', 'Liquidificador'],
    'Vestuário': ['Camiseta', 'Vestido', 'Jaqueta', 'Calça', 'Blusa']
}

lista = []

for i in range(101):
  nome = pessoa_random().nome
  categoria = random.choice(list(categorias_e_produtos.keys()))
  produto = random.choice(list(categorias_e_produtos[categoria]))
  data_compra = Faker().date()
  quantidade = random.randint(1, 20)
  preco = round(random.uniform(30, 2000), 2)

  lista.append([nome, produto ,categoria, data_compra, quantidade, preco])

df = pd.concat([df, pd.DataFrame(lista, columns=df.columns)], ignore_index=True)

df
```

```python
df.to_csv('vendas.csv', index=False, encoding='utf-8')
```
